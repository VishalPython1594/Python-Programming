## Shallow Copy in List

- Both the list are Different
- Any change in one list will NOT be executed in the other 
- any change in sublist of the parent list will be executed in the sublist of the copied list also

**Syntax : listName.copy()**  List Method

**Syntax : copy.copy(listName)**  Package copy -> copy method


```python
listA = ["python" , "java" , ['ruby' , 'php' , 3 , 4] , 10.3 , False]
listB = listA.copy()

print(f"Memory address of listA : {id(listA)}")
print(f"Memory address of listB : {id(listB)}")
```

```python
listA = ["python" , "java" , ['ruby' , 'php' , 3 , 4] , 10.3 , False]
listB = listA.copy()  # SHALLOW COPY

print(f"listA ==> {listA}")
print(f"listB ==> {listB}")
print() 

print(f"Memory address of listA : {id(listA)}")
print(f"Memory address of listB : {id(listB)}")
print() 

print("Performing Action on listA : pop()")

listA.pop()
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")


print()
print("Performing Action on listB : append()")

listB.append(99)
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")


print()
print("Performing Action on listA sublist : remove('php')")

listA[2].remove('php')
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")


print()
print("Performing Action on listB sublist : pop()")

listB[2].pop()
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")
```

## Deep Copy in List

- Both the list are Different
- Any change in one list will NOT be executed in the other 
- any change in sublist of the parent list will NOT be executed in the sublist of the copied list also

**Syntax : copy.deepcopy(listName)**  Package copy -> copy method


```python
import copy 

listA = ["python" , "java" , ['ruby' , 'php' , 3 , 4] , 10.3 , False]
listB = copy.deepcopy(listA)  # DEEP COPY

print(f"listA ==> {listA}")
print(f"listB ==> {listB}")
print() 

print(f"Memory address of listA : {id(listA)}")
print(f"Memory address of listB : {id(listB)}")
print() 

print("Performing Action on listA : pop()")

listA.pop()
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")


print()
print("Performing Action on listB : append()")

listB.append(99)
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")


print()
print("Performing Action on listA sublist : remove('php')")

listA[2].remove('php')
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")


print()
print("Performing Action on listB sublist : pop()")

listB[2].pop()
print(f"listA ==> {listA}")
print(f"listB ==> {listB}")
```

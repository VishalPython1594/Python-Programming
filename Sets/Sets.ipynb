## Set Methods


```python
a = {"@", "#"}
b = {1 , 2 , 3}
c = {1 , 3, "@"}
```

```python
# SUBSET and SUPERSET

a = {"@", "#"}
b = {1 , 2 , 3}
c = {1 , 3, "@"}
d = {1 , 2 , 3, "@", "#"}


print( d.issuperset(a) )
print( a.issubset(d) )

print( a.issubset(c) )

```

```python
# DISJOINT

a = {"@", "#"}
b = {1 , 2 , 3}
c = {1 , 3, "@"}
d = {1 , 2 , 3, "@", "#"}

print( c.isdisjoint(a)  )
print( b.isdisjoint(a)  )  
# print( a.isdisjoint(b)  )

```

```python
# UNION

a = {"@", "#"}
b = {1 , 2 , 3}
c = {1 , 3, "@"}
d = {1 , 2 , 3, "@", "#"}

print( a.union(b) )
print( b.union(c) )
```

```python
# INTERSECTION

a = {"@", "#"}
b = {1 , 2 , 3}
c = {1 , 3, "@"}
d = {1 , 2 , 3, "@", "#"}

print( a.intersection(b) )
print( b.intersection(c) )
```



### B - A

```python
# DIFFERENCE

# ans = set_1 - (set_1 intersection set_2)
# PURE SET / WITHOUT ANY COMMON 

a = {"@", "#"}
b = {1 , 2 , 3}
c = {1 , 3, "@"}
d = {1 , 2 , 3, "@", "#"}

print( a.difference(b) )
print( c.difference(b) ) # C - B

print( d.difference(b) ) # D - B
print( b.difference(d) ) # B - D

```

# Set Mutable Methods

```python
# add : Insert SINGLE element 

setName = {1 , 2 , 3, "@", "#"}

setName.add(99)
print(setName)


setName.add(False)
print(setName)


setName.add('ruby')
print(setName)


setName.add((-3 , -4)  )
print(setName)

setName.add(-3 , -4  )
print(setName)
```

```python
# update : Insert MUTLIPLE element 

setName = {1 , 2 , 3, "@", "#"}

setName.update([1 , 1 , '&', ('p', 'w')])
# setName.update(1 , 1 , '&', ('p', 'w'))
print(setName)

```

```python
# UPDATE XX

setName = {1 , 2 , 3, "@", "#"}
setName[0] = 99
```

```python
# REMOVE - this will remove the element by referring to its value

setName = {1 , 2 , 3, "@", "#"}
setName.remove(2) # remove by VALUE
print( setName )

setName.remove(2) # remove by VALUE
print( setName )

```

```python
# POP - this will remove element by referring to its position RANDMLY

setName = {1 , 2 , 3, "@", "#"}
setName.pop()
print( setName )

setName.pop()
print( setName )

setName.pop()
print( setName )
```

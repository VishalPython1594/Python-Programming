```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Univariate Analysis -- 
# Single column
# Numerical column -> histogram
# Categorical vs Numerical(count) column -> bar, pie, area, heatmap, donut...

# Bivariate Analysis -- 
# 2 column
# Numerical vs Numerical column -> scatter, line
# **Categorical vs Numerical column -> bar, pie, area, heatmap, donut...


# Multi-ivariate Analysis -- compare more than 2 columns
# 3 column : represent it as a color
# Numerical vs Numerical (vs Categorical) column -> scatter, line
# **Categorical vs Numerical (vs Categorical) column -> bar, pie, area, heatmap, donut...

```

```python
# Univariate Analysis -- 
# Single column
# Categorical vs Numerical(count) column -> bar, pie, area, heatmap, donut...
```

```python
gender_group = df.groupby('Sex').aggregate( totalRecords = ("Sex", np.size) )
gender_group
```

```python
x_ = gender_group.index.values
print(x_)

y_ = gender_group['totalRecords'].values
print(y_)
```

```python
plt.bar( x = gender_group.index.values, height= gender_group['totalRecords'].values, color = 'm')
plt.xlabel("Sex")
plt.ylabel("frequency")
plt.title("--- Count Chart ---")
plt.show()
```

```python
plt.bar( x = gender_group.index.values, 
        height= gender_group['totalRecords'].values, 
        color = ['orange', "red"])
plt.xlabel("Sex")
plt.ylabel("frequency")
plt.title("--- Count Chart ---")
plt.show()
```

```python
gender_group = df.groupby('Sex').aggregate( totalRecords = ("Sex", np.size) )
gender_group = gender_group.sort_values(by = "totalRecords", ascending = False)


plt.bar( x = gender_group.index.values, 
        height= gender_group['totalRecords'].values, 
        color = ['orange', "red"])
plt.xlabel("Sex")
plt.ylabel("frequency")
plt.title("--- Count Chart ---")
plt.show()
```

```python
# Bivariate Analysis -- 
# 2 column
# **Categorical vs Numerical column -> bar, pie, area, heatmap, donut...
```

```python
Embarked_group = df.groupby('Embarked').aggregate( totalRecords = ("Embarked", np.size), 
                                                 avg_age = ("Age", np.mean), 
                                                 totalFare = ("Fare", np.sum))
Embarked_group
```

```python
plt.pie(x = Embarked_group['avg_age'].values,
    labels=Embarked_group.index.values, labeldistance = 0.5, autopct = "%0.2f%%",
    colors=['springgreen', "orchid", "navy"])

plt.title("--- embark vs Age Pie Chart ---")
plt.show()
```

```python
plt.figure( figsize = (20 , 10))

plt.subplot( 1, 3 , 1 )
plt.pie(x = Embarked_group['avg_age'].values,
    labels=Embarked_group.index.values, autopct = "%0.2f%%", 
    colors=['springgreen', "orchid", "orange"])
plt.title("Age Analysis")

plt.subplot( 1, 3 , 2 )
plt.pie(x = Embarked_group['totalFare'].values,
    labels=Embarked_group.index.values,autopct = "%0.2f%%", 
    colors=['springgreen', "orchid", "orange"])
plt.title("Fare Analysis")



plt.subplot( 1, 3 , 3 )
plt.pie(x = Embarked_group['totalRecords'].values,
    labels=Embarked_group.index.values,autopct = "%0.2f%%", 
    colors=['springgreen', "orchid", "orange"])
plt.title("Count Analysis")



plt.show()
```

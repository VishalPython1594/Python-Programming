## NA and NULL Values

NULL -> 
- Absence of Value
- Its not countable


NA -> superset of NULL
- Presence of Value (Uncertain value)
- Its countable


```python
[np.nan, 10]
```

```python
print( np.nan + 10 )
print( np.nan / 10 )
```

### Filling NA values inside the column


**dfName[columnName].fillna(value = , inplace = True)**


```python
myTable_NA['A1'].mean()
```

```python
#### Fill NA values with Mean in A1
```

```python
myTable_NA['A1'].fillna( value= myTable_NA['A1'].mean(), inplace=True )
```

```python
myTable_NA
```

```python
#### Fill NA values with Median in A3
```

```python
myTable_NA['A3'].fillna( value= myTable_NA['A3'].median(), inplace=True )
```

```python
myTable_NA
```

```python
print( myTable_NA['A4'].mode() ) # return all the values

print( myTable_NA['A4'].mode()[0] ) # return FIRST values
```

```python
myTable_NA['A4'].fillna( value=myTable_NA['A4'].mode()[0], inplace=True )
```

```python
myTable_NA
```

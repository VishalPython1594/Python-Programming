#pd.merge(left, right, on, how)

left = left table right = right table how = type of join ==> inner, outer, left, right

on = common column Nmae / Primary Key

OR,

left_on -> left column name right_on -> right column name

```python
d1 = pd.read_excel('datasets/studentCollegeData.xlsx',sheet_name='collegeType')
d2 = pd.read_excel('datasets/studentCollegeData.xlsx',sheet_name='regions')
d1.head()
```

```python
d2.head()
```

```python
left_join = pd.merge(left=d1,right=d2,how='left',on='School Name')
print(left_join)
left_join.shape
```

```python
left_join1 = pd.merge(left=d2,right=d1,how='left',on='School Name')
print(left_join1)
left_join1.shape
```
